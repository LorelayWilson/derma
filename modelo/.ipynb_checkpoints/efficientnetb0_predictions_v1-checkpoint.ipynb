## 1. IMPORTACIN DE LIBRERAS
Importamos las librer铆as necesarias para el procesamiento de datos, construcci贸n y entrenamiento del modelo.

```python
import os
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import mixed_precision
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
```

## 2. Configuraci贸n de par谩metros y recursos
Configuramos los par谩metros principales y verificamos la disponibilidad de GPU para acelerar el entrenamiento.

```python
mixed_precision.set_global_policy('float32')

# Par谩metros
img_size = (224, 224)
batch_size = 32
epochs = 100
seed = 123
train_dir = '../dataset/multi-6/train'
test_dir = '../dataset/multi-6/test'
model_path = 'efficientnetb0_derma_v5-2.keras'
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="training",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    validation_split=0.2,
    subset="validation",
    seed=seed,
    image_size=img_size,
    batch_size=batch_size
)
# Obtener n煤mero de clases
class_names = train_ds.class_names
num_classes = len(class_names)
print("Clases:", class_names)

#Imprimir dataset
for x, y in train_ds.take(5):
    print("Imagen shape:", x.shape)
    print("Etiqueta:", y.numpy())
```

```python
from tensorflow.keras.utils import register_keras_serializable

#@register_keras_serializable()
def cost_matrix_sparse_categorical_crossentropy_loss(y_true, y_pred, alpha=1.0):
    # Aseg煤rate de usar num_classes correcto si no es global
    num_classes = tf.shape(y_pred)[-1]

    # Definir clases por nivel de riesgo
    high_danger = {4}  # Malign
    moderate_danger = {2, 3}  # Eczema, Infectious
    all_classes = set(range(5))  # Actualiza seg煤n num_classes
    low_danger = all_classes - high_danger - moderate_danger

    # Construir la matriz de costos (en cada llamada, por ahora)
    cost_matrix = tf.ones([num_classes, num_classes], dtype=tf.float32)
    cost_matrix = tf.where(tf.eye(num_classes) == 1, 0.0, cost_matrix)

    for high in high_danger:
        for low in low_danger:
            cost_matrix = tf.tensor_scatter_nd_update(
                cost_matrix, indices=[[high, low]], updates=[10.0])
        for mid in moderate_danger:
            cost_matrix = tf.tensor_scatter_nd_update(
                cost_matrix, indices=[[high, mid]], updates=[5.0])

    for mid in moderate_danger:
        for low in low_danger:
            cost_matrix = tf.tensor_scatter_nd_update(
                cost_matrix, indices=[[mid, low]], updates=[5.0])

    # y_true es sparse (enteros), convertir a one-hot
    y_true_oh = tf.one_hot(tf.cast(tf.reshape(y_true, [-1]), tf.int32), depth=num_classes)

    # Coste esperado por fila
    row_cost = tf.matmul(y_true_oh, cost_matrix)

    # Penalizaci贸n por predicci贸n
    y_pred_probs = tf.nn.softmax(y_pred)
    exp_cost = tf.reduce_sum(row_cost * y_pred_probs, axis=-1)

    # Cross entropy normal
    ce = tf.keras.losses.categorical_crossentropy(y_true_oh, y_pred_probs)

    return ce + alpha * exp_cost
```

## 8. Evaluaci贸n y visualizaci贸n de resultados
Evaluamos el modelo en el conjunto de test, mostramos el reporte de clasificaci贸n y la matriz de confusi贸n.

```python

# Cargar modelo entrenado
model = tf.keras.models.load_model(model_path, custom_objects={'cost_matrix_sparse_categorical_crossentropy_loss': cost_matrix_sparse_categorical_crossentropy_loss})

num_classes = len(class_names)

y_pred_probs = model.predict(val_ds)
y_pred = np.argmax(y_pred_probs, axis=1)

y_true = np.concatenate([y.numpy() for x, y in val_ds])

print("\nЬ Clasification Report:\n")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(14, 10))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicci贸n')
plt.ylabel('Real')
plt.title('Matriz de Confusi贸n')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Las 3 clases m谩s probables

```python
from IPython.display import display, HTML

# Definir el nivel de gravedad de cada clase
gravedad = {
    0: 0,  # Acne
    1: 0,  # Benign
    2: 1,  # Eczema
    3: 1,  # Infectious
    4: 2,  # Malign
    5: 0   # Pigmentation
}

top_k = 4  # Considerar las 4 clases m谩s probables
num_imgs = 50  # Primeras 10 im谩genes

y_true = np.concatenate([y.numpy() for x, y in val_ds])

for i in range(num_imgs):
    probs = y_pred_probs[i]
    real_idx = y_true[i]
    top_indices = np.argsort(probs)[-top_k:][::-1]
    idx1 = top_indices[0]
    conf1 = probs[idx1]
    pred_idx = idx1
    motivo = ""
    diferencia_elegida = 0.0
    
    # Buscar entre las top_k si hay alguna m谩s grave y cercana en confianza
    for idx in top_indices[1:]:
        conf = probs[idx]
        diff = conf1 - conf
        if (diff < 0.10) and (gravedad[idx] > gravedad[pred_idx]):
            pred_idx = idx
            motivo = f"<span style='color:orange'>*Se elige {class_names[idx]} por mayor gravedad y confianza cercana*</span>"
            diferencia_elegida = diff

    color = "green" if pred_idx == real_idx else "red"

    html = f"<b>Imagen {i+1}:</b><br>"
    for j, idx in enumerate(top_indices):
        diff = conf1 - probs[idx]
        if j == 0:
            html += f"{j+1}陋: {class_names[idx]}: {probs[idx]*100:.2f}%<br>"
        else:
            html += f"{j+1}陋: {class_names[idx]}: {probs[idx]*100:.2f}% (diferencia: {diff*100:.2f}%)<br>"
    if motivo:
        html += motivo + f" (diferencia: {diferencia_elegida*100:.2f}%)<br>"
    html += f"<b>Predicci贸n final:</b> <span style='color:{color}'>{class_names[pred_idx]}</span><br>"
    html += f"<b>Real:</b> {class_names[real_idx]}"
    html += "<hr>"
    display(HTML(html))
```
